```python
#comlete the from num2words import num2words
import re

def convert_numbers_to_words(text):
    def replace(match):
        return num2words(int(match.group()))
    return re.sub(r'\d+', replace, text)
```

```python
print(convert_numbers_to_words("I have 2 apples and 15 bananas"))
```

![ ](https://img.shields.io/badge/drop_stop_words%20%20-blue)

```python
def remove_stopwords(text):
    doc = nlp(text)
    result = []
    for token in doc:
        if not token.is_stop:
            result.append(token.text)
    return " ".join(result)
```

```python
text = "This is a simple example to show how stopwords are removed"
print(remove_stopwords(text))
```

![ ](https://img.shields.io/badge/drop_punctuations%20%20-blue)

```python
def remove_punctuation(text):
    result = ""
    for char in text:
        if char not in punctuations:
            result += char
    return result
```

```python
text = "Hello!!! How are you??? I am fine :) #professeur sahrauoi "
print(remove_punctuation(text))
```

![ ](https://img.shields.io/badge/Process%20%20-orange)

```python
def process_text(text):
    text = text.lower()
    text = convert_numbers_to_words(text)
    text = remove_punctuation(text)
    text = remove_stopwords(text)
    return text

processed_X = []
for doc in X:
    processed_X.append(process_text(doc))
```

```python
test = "I have 2 apples, and this is a simple test!!!"
print(process_text(test))
```

![ ](https://img.shields.io/badge/Vocabs%20%20-blue)

```python
# vocab
vocab = []

for text in processed_X:
    words = text.split()
    for w in words:
        if w not in vocab:
            vocab.append(w)

# vectors
def text_to_vector(text, vocab):
    words = text.split()
    vector = [0] * len(vocab)
    
    for w in words:
        if w in vocab:
            index = vocab.index(w)
            vector[index] = 1
    
    return vector

Vectors = []
for text in processed_X:
    Vectors.append(text_to_vector(text, vocab))
```
