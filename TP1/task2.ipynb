```python
# =========================
#DATA
# =========================

docs = [
    "Muslims are praying and giving charity today",
    "People are fasting and praying during Ramadan",
    "The community gathers to pray and help the needy",
    "Customers are buying and giving feedback today",
    "People are shopping and giving reviews online",
    "The community gathers to buy and help new clients"
]

labels = [0,0,0,1,1,1]

# =========================
#  CLEANING + NORMALIZATION
# =========================

punctuation = ".,!?;:"
stopwords = ["are", "and", "to", "the", "during"]

def preprocess(text):
    # lowercase
    text = text.lower()
    
    # remove punctuation
    clean = ""
    for c in text:
        if c not in punctuation:
            clean += c
    
    # tokenization
    words = clean.split()
    
    # remove stopwords
    final_words = []
    for w in words:
        if w not in stopwords:
            final_words.append(w)
    
    return final_words


processed_docs = [preprocess(doc) for doc in docs]

print(processed_docs)
```

```python
# =========================
# VOCABULARY
# =========================

vocab = []

for doc in processed_docs:
    for word in doc:
        if word not in vocab:
            vocab.append(word)

print("\nVocabulary:")
print(vocab)

# =========================
# VECTORIZATION
# =========================

def vectorize(doc, vocab):
    vector = []
    for word in vocab:
        if word in doc:
            vector.append(1)
        else:
            vector.append(0)
    return vector

X = [vectorize(doc, vocab) for doc in processed_docs]

print("\nVectors:")
for v in X:
    print(v)
```

```python
# =========================
# KMEANS
# =========================

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

print("\nClusters:")
print(kmeans.labels_)
```

```python
The main problem with one-gram Bag of Words is that it ignores the order and context of words.

For example:
"buy product" and "product buy" have the same representation.

Also, it cannot understand meaning:
"not good" and "good" may be treated similarly.

Therefore, BOW loses semantic information.
```
